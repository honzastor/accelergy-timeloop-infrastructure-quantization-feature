## Processing of experiments result data

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import glob
import os
import math
import re
```

```python
# Helper functions
def reduce_nesting_level(path, levels):
    """
    Recursively reduces the nesting level of a path by `levels`.

    Args:
        path (str): The path to reduce the nesting level of.
        levels (int): The number of levels to reduce.

    Returns:
        str: The reduced path.
    """
    if levels == 0:
        return path

    return reduce_nesting_level(os.path.dirname(path), levels - 1)

def extract_utilized_capacity(filename):
    """
    Extracts the utilized memory capacity from a stats.txt file.

    Args:
        filename (str): The path to the file to extract the utilized capacity from.

    Returns:
        int: Total utilized memory capacity.
    """
    with open(filename, 'r') as file:
        content = file.read()

    pattern = r"Utilized capacity\s+: (\d+)"
    utilized_capacity_values = re.findall(pattern, content)

    utilized_capacity_sum = sum(int(value) for value in utilized_capacity_values)
    return utilized_capacity_sum


def extract_utilized_capacity_weights(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

    utilized_capacities = []
    levels = content.split('Level ')
    for level in levels:
        if 'Weights:' in level:
            match = re.search('Utilized capacity\s*:\s*(\d+)', level)
            if match:
                utilized_capacities.append(int(match.group(1)))

    return sum(utilized_capacities)
```

## Experiment 1: Valid mappings scatterplot (Eyeriss_like; AlexNet; w and w/o gating)
* Scatter and Box plots of individual valid mapping results for individual data quantization settings for chosen Eyeriss desing (with and without gating optimization) and chosen AlexNet CONV layer
* The considered scenario is found in the `../../scenarios/alexnet_eyeriss-like/mapper` folder
* You can change the desired `metric`, `design_setting` and `layer_choice` in the code

```python
# Scenario path
scenario = "../../scenarios/alexnet_eyeriss-like/mapper"
# Customizable parameters
metric = "edp" # edp, energy-delay, delay-energy, energy-last-level-accesses, last-level-accesses-energy, delay-last-level-accesses, last-level-accesses-delay

### RETRIEVE DATA ###
path = os.path.join(scenario)
arch_name = scenario.split("/")[3].split("_")[1].split("-")[0]

# Define column names for the dataframe (OAVES data are typically collected by multiple threads and thus do not have a header row)
column_names = [
    "pJ/Compute",
    "Utilization",
    "Cycles",
    "Energy [uJ]",
    "EDP [J*cycle]",
    "Area [mm^2]",
    "Computes",
    "Algorithmic Computes",
    "Buffer Size",
    "Op Intensity",
    "LastLevelAccesses",
    "Buffer Size Tensor 1",
    "Buffer Size Tensor 2",
    "Buffer Size Tensor 3",
    "DRAM Word Accesses Tensor 1",
    "DRAM Word Accesses Tensor 2",
    "DRAM Word Accesses Tensor 3",
    "Compact Print of Mapping",
    "Path to Mapping.yaml"
]

oaves_dfs = [] # oaves.csv data
stats_dfs = [] # stats.csv data
# Extract results from the Eyeriss_like architecture's oaves.csv
for f in glob.glob(f"{path}/**/{metric}/**/*oaves.csv", recursive=True):
    design_setting = os.path.basename(reduce_nesting_level(f, 8))
    quant_setting = os.path.basename(reduce_nesting_level(f, 7))
    mapper_name = os.path.basename(reduce_nesting_level(f, 3))
    metric_name = os.path.basename(reduce_nesting_level(f, 2))
    topology_name = os.path.basename(reduce_nesting_level(f, 1))
    layer_name = os.path.splitext(os.path.splitext(os.path.basename(f))[0])[0]
    # print(f"arch_name: {arch_name}, design_setting: {design_setting}, quant_setting: {quant_setting}, mapper_name: {mapper_name}, metric_name: {metric_name}, topology_name: {topology_name}, layer_name: {layer_name}")
    
    arch_df = pd.read_csv(f, names=column_names)
    arch_df['Architecture'] = arch_name
    arch_df['NN topology'] = topology_name
    arch_df['Layer'] = layer_name
    arch_df['Mapper'] = mapper_name
    arch_df['Metric'] = metric_name
    arch_df['Datawidth setting'] = quant_setting
    arch_df['Design setting'] = design_setting
    oaves_dfs.append(arch_df)
oaves_df = pd.concat(oaves_dfs, axis=0, ignore_index=True)

# Extract results from the Eyeriss_like architecture's stats.csv
for f in glob.glob(f"{path}/**/{metric}/**/*stats.csv", recursive=True):
    design_setting = os.path.basename(reduce_nesting_level(f, 8))
    quant_setting = os.path.basename(reduce_nesting_level(f, 7))
    mapper_name = os.path.basename(reduce_nesting_level(f, 3))
    metric_name = os.path.basename(reduce_nesting_level(f, 2))
    topology_name = os.path.basename(reduce_nesting_level(f, 1))
    layer_name = os.path.splitext(os.path.splitext(os.path.basename(f))[0])[0]
    # print(f"arch_name: {arch_name}, design_setting: {design_setting}, quant_setting: {quant_setting}, mapper_name: {mapper_name}, metric_name: {metric_name}, topology_name: {topology_name}, layer_name: {layer_name}")
    
    arch_df = pd.read_csv(f)
    arch_df['Architecture'] = arch_name
    arch_df['NN topology'] = topology_name
    arch_df['Layer'] = layer_name
    arch_df['Mapper'] = mapper_name
    arch_df['Metric'] = metric_name
    arch_df['Datawidth setting'] = quant_setting
    arch_df['Design setting'] = design_setting
    stats_dfs.append(arch_df)
stats_df = pd.concat(stats_dfs, axis=0, ignore_index=True)
```

```python
""" SCATTER PLOT """
### MODIFY/SELECT DATA ###
design_setting = "no_optimizations" # gating_optimization, no_optimizations
layer_choice = 'AlexNet_layer3' # AlexNet_layer1, AlexNet_layer2, AlexNet_layer3, AlexNet_layer4, AlexNet_layer5

scatter_plot_df = oaves_df.loc[(oaves_df['Layer'] == layer_choice) & (design_setting == oaves_df['Design setting'])]
best_mappings_df = stats_df.loc[(stats_df['Layer'] == layer_choice) & (design_setting == stats_df['Design setting'])]

quant_settings = scatter_plot_df['Datawidth setting'].unique()

# Find the best mapping for each quantization setting
scatter_plot_df.loc[:, 'Best'] = False
for quant_setting in quant_settings:
    quant_setting_df = scatter_plot_df[scatter_plot_df['Datawidth setting'] == quant_setting]
    best_mapping = quant_setting_df.loc[quant_setting_df['EDP [J*cycle]'].idxmin()]
    scatter_plot_df.loc[best_mapping.name, 'Best'] = True

### PLOT GRAPHS ###
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 5), sharex=True, sharey=True)

for i, setting in enumerate(quant_settings):
    sub_df = scatter_plot_df[scatter_plot_df['Datawidth setting'] == setting]
    best_df = sub_df[sub_df['Best']]
    other_df = sub_df[~sub_df['Best']]
    axes[i].scatter(other_df['Cycles'], other_df['Energy [uJ]'], s=20, alpha=0.2, c='tab:blue', rasterized=True)
    axes[i].scatter(best_df['Cycles'], best_df['Energy [uJ]'], s=40, alpha=1, c='r', marker='x', rasterized=True)    
    axes[i].set_xlabel('Cycles', fontsize=16, fontweight='bold')
    if i == 0:
        axes[i].set_ylabel('Energy [uJ]', fontsize=16, fontweight='bold')
    axes[i].set_title(setting, fontsize=16)
    axes[i].grid(True, color='lightgray', linestyle='-', linewidth=0.5, alpha=0.25)
    axes[i].tick_params(axis='both', which='major', labelsize=14)

#saving the graph image
plt.tight_layout()
fig_location = "plots/Experiment1"
if not os.path.exists(fig_location):
    os.makedirs(fig_location)
plt.savefig(os.path.join(fig_location, f'{arch_name}_{design_setting}_{layer_choice}_{metric}_scatter.png'), bbox_inches='tight')
plt.savefig(os.path.join(fig_location, f'{arch_name}_{design_setting}_{layer_choice}_{metric}_scatter.pdf'), bbox_inches='tight')

plt.suptitle(f"{arch_name} {design_setting} {layer_choice} {metric} optimized metric", fontsize=18, y=1.05, x=0.53)
plt.show()
```

```python
# Get the total number of records for each quant_setting (NOTE that the numbers do not represent total number of valid mappings per setting! Because just the best evaluated mappings per thread is reported (in other words, the number is probably less than the actual number of valid mappings and will change with different number of threads used, to see true counts, run the mapper with live-status option on))
total_records = scatter_plot_df.groupby(['Datawidth setting']).size().reset_index(name='Total Records')
print(total_records.sort_values('Total Records'))

for quant_setting in quant_settings:
    quant_setting_df = scatter_plot_df[scatter_plot_df['Datawidth setting'] == quant_setting]
    best_mapping = quant_setting_df.loc[quant_setting_df['EDP [J*cycle]'].idxmin()]
    scatter_plot_df.loc[best_mapping.name, 'Best'] = True
    print(f"{quant_setting} best EDP mapping: {best_mapping['EDP [J*cycle]']} J*cycle")
```

```python
# Scenario path
scenario = "../../scenarios/trained_mobilenet_eval"
### RETRIEVE DATA ###
path = os.path.join(scenario)

""" extract stats.csv data """

native_stats_dfs = [] # original float model stats.csv data
quant_stats_dfs = []  # quantized int model stats.csv data
# Extract results from the native model's stats.csv
for f in glob.glob(f"{path}/**/original_float_data/**/*stats.csv", recursive=True):
    arch_name = os.path.basename(reduce_nesting_level(f, 9)).split('-')[0]
    mapper_name = os.path.basename(reduce_nesting_level(f, 5))
    metric_name = os.path.basename(reduce_nesting_level(f, 4))
    design_setting = os.path.basename(reduce_nesting_level(f, 3))
    topology_name = os.path.basename(reduce_nesting_level(f, 2))
    quant_setting = os.path.basename(reduce_nesting_level(f, 1))
    layer_name = os.path.splitext(os.path.splitext(os.path.basename(f))[0])[0]
    # print(f"arch_name: {arch_name}, design_setting: {design_setting}, quant_setting: {quant_setting}, mapper_name: {mapper_name}, metric_name: {metric_name}, topology_name: {topology_name}, layer_name: {layer_name}")
       
    arch_df = pd.read_csv(f)
    arch_df['Architecture'] = arch_name
    arch_df['Design setting'] = design_setting
    arch_df['NN topology'] = topology_name
    arch_df['Layer'] = layer_name
    arch_df['Mapper'] = mapper_name
    arch_df['Metric'] = metric_name
    arch_df['Datawidth setting'] = quant_setting
    native_stats_dfs.append(arch_df)  
native_stats_df = pd.concat(native_stats_dfs, axis=0, ignore_index=True)

# Extract results from the quantized model's stats.csv
for f in glob.glob(f"{path}/**/quantized_data/**/*stats.csv", recursive=True):
    arch_name = os.path.basename(reduce_nesting_level(f, 10)).split('-')[0]
    mapper_name = os.path.basename(reduce_nesting_level(f, 5))
    metric_name = os.path.basename(reduce_nesting_level(f, 4))
    design_setting = os.path.basename(reduce_nesting_level(f, 3))
    topology_name = os.path.basename(reduce_nesting_level(f, 2))
    quant_setting = os.path.basename(reduce_nesting_level(f, 1))
    layer_name = os.path.splitext(os.path.splitext(os.path.basename(f))[0])[0]
    # print(f"arch_name: {arch_name}, design_setting: {design_setting}, quant_setting: {quant_setting}, mapper_name: {mapper_name}, metric_name: {metric_name}, topology_name: {topology_name}, layer_name: {layer_name}")
       
    arch_df = pd.read_csv(f)
    arch_df['Architecture'] = arch_name
    arch_df['Design setting'] = design_setting
    arch_df['NN topology'] = topology_name
    arch_df['Layer'] = layer_name
    arch_df['Mapper'] = mapper_name
    arch_df['Metric'] = metric_name
    arch_df['Datawidth setting'] = quant_setting
    quant_stats_dfs.append(arch_df)
quant_stats_df = pd.concat(quant_stats_dfs, axis=0, ignore_index=True)


""" extract stats.txt data """

native_util_cap_dfs = [] # original float model stats.csv data
quant_util_cap_dfs = []  # quantized int model stats.csv data
# Extract results from the native model's stats.txt
for f in glob.glob(f"{path}/**/original_float_data/**/*stats.txt", recursive=True):
    arch_name = os.path.basename(reduce_nesting_level(f, 9)).split('-')[0]
    mapper_name = os.path.basename(reduce_nesting_level(f, 5))
    metric_name = os.path.basename(reduce_nesting_level(f, 4))
    design_setting = os.path.basename(reduce_nesting_level(f, 3))
    topology_name = os.path.basename(reduce_nesting_level(f, 2))
    quant_setting = os.path.basename(reduce_nesting_level(f, 1))
    layer_name = os.path.splitext(os.path.splitext(os.path.basename(f))[0])[0]
    # print(f"arch_name: {arch_name}, design_setting: {design_setting}, quant_setting: {quant_setting}, mapper_name: {mapper_name}, metric_name: {metric_name}, topology_name: {topology_name}, layer_name: {layer_name}")

    arch_df = pd.DataFrame()
    arch_df['Architecture'] = arch_name
    arch_df['Design setting'] = design_setting
    arch_df['NN topology'] = topology_name
    arch_df['Layer'] = layer_name
    arch_df['Mapper'] = mapper_name
    arch_df['Metric'] = metric_name
    arch_df['Datawidth setting'] = quant_setting
    arch_df['Utilized capacity'] = extract_utilized_capacity(f)
    arch_df['Utilized weight capacity'] = extract_utilized_capacity_weights(f)
    native_util_cap_dfs.append(arch_df)
native_util_cap_df = pd.concat(native_util_cap_dfs, axis=0, ignore_index=True)
native_stats_df['Utilized capacity'] = native_util_cap_df['Utilized capacity']
native_stats_df = pd.merge(native_stats_df, native_util_cap_df[['Architecture', 'Design setting', 'NN topology', 'Layer', 'Mapper', 'Metric', 'Datawidth setting', 'Utilized capacity', 'Utilized weight capacity']], on=['Architecture', 'Design setting', 'NN topology', 'Layer', 'Mapper', 'Metric', 'Datawidth setting'], how='left')


# Extract results from the quantized model's stats.txt
for f in glob.glob(f"{path}/**/quantized_data/**/*stats.txt", recursive=True):
    arch_name = os.path.basename(reduce_nesting_level(f, 10)).split('-')[0]
    mapper_name = os.path.basename(reduce_nesting_level(f, 5))
    metric_name = os.path.basename(reduce_nesting_level(f, 4))
    design_setting = os.path.basename(reduce_nesting_level(f, 3))
    topology_name = os.path.basename(reduce_nesting_level(f, 2))
    quant_setting = os.path.basename(reduce_nesting_level(f, 1))
    layer_name = os.path.splitext(os.path.splitext(os.path.basename(f))[0])[0]
    # print(f"arch_name: {arch_name}, design_setting: {design_setting}, quant_setting: {quant_setting}, mapper_name: {mapper_name}, metric_name: {metric_name}, topology_name: {topology_name}, layer_name: {layer_name}")
       
    arch_df = pd.read_csv(f)
    arch_df['Architecture'] = arch_name
    arch_df['Design setting'] = design_setting
    arch_df['NN topology'] = topology_name
    arch_df['Layer'] = layer_name
    arch_df['Mapper'] = mapper_name
    arch_df['Metric'] = metric_name
    arch_df['Datawidth setting'] = quant_setting
    arch_df['Utilized capacity'] = extract_utilized_capacity(f)
    arch_df['Utilized weight capacity'] = extract_utilized_capacity_weights(f)
    quant_util_cap_dfs.append(arch_df)
quant_util_cap_df = pd.concat(quant_util_cap_dfs, axis=0, ignore_index=True)
quant_stats_df = pd.merge(quant_stats_df, quant_util_cap_df[['Architecture', 'Design setting', 'NN topology', 'Layer', 'Mapper', 'Metric', 'Datawidth setting', 'Utilized capacity', 'Utilized weight capacity']], on=['Architecture', 'Design setting', 'NN topology', 'Layer', 'Mapper', 'Metric', 'Datawidth setting'], how='left')
```

```python
" SCATTER PLOT OF RELATIVE ACHIEVED DESIRED METRIC "
### MODIFY/SELECT DATA ###
# "original_float32": 61.40
data = {
    "act8_weights2": 10.00,
    "act8_weights3": 54.59,
    "act8_weights4": 58.14,
    "act8_weights5": 61.02,
    "act8_weights6": 61.88,
    "act8_weights7": 62.05,
    "act8_weights8": 62.39
}

### MODIFY/SELECT DATA ###
arch_name = "eyeriss"    # "eyeriss", "simba"
# CYCLES ARE HIGHER FOR CONFIGURATIONS WITH SMALLER DATAWIDTHS PROBABLY DUE TO EDP BEING THE OPTIMIZED METRIC (FOUND MAPINGS WITH LESS ENERGY AND OVERALL BETTER EDP EVEN THOUGH CYCLES ARE HIGHER)!
metric = "Utilized capacity" # Cycles, EDP [J*cycle], Energy [uJ], Utilization, pJ/Compute, Area [mm^2], Computes, Algorithmic Computes, Utilized capacity, Utilized weight capacity

# Create DataFrame from the data
quant_df_tmp = quant_stats_df[(quant_stats_df['Architecture'] == arch_name)] 
scatter_plot_df = quant_df_tmp.groupby(['Design setting', 'Datawidth setting']).sum().reset_index()


### PLOT GRAPHS ###
# Set up the subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(16, 6), sharey=True)

# Get unique Design settings
design_settings = scatter_plot_df["Design setting"].unique()

# Get unique Datawidth settings for the current Design setting
datawidth_settings = scatter_plot_df["Datawidth setting"].unique()
    
# Create a color map with the number of unique datawidth settings
cmap = plt.cm.get_cmap('tab10', len(datawidth_settings))

# Initialize a dictionary to store the normalized values
normalized_values_dict = {k: {} for k in design_settings}

# Iterate over Design settings
for i, design_setting in enumerate(design_settings):
    # Filter DataFrame for the current Design setting
    sub_df = scatter_plot_df[scatter_plot_df["Design setting"] == design_setting]
    
    
    # Initialize lists to store x and y values for scatter plot
    x_values = []
    y_values = []
    c_values = []
    
    # Get the reference value for normalization ('act8_weights8')
    reference_value = sub_df.loc[sub_df['Datawidth setting'] == 'act8_weights8', metric].values[0]
    
    # Iterate over Datawidth settings
    for j, datawidth_setting in enumerate(datawidth_settings):
        # Filter DataFrame for the current Datawidth setting
        datawidth_df = sub_df[sub_df["Datawidth setting"] == datawidth_setting]
        
        # Get the accuracy value for the current Datawidth setting
        accuracy = data[datawidth_setting]
        
        # Calculate normalized latency value
        normalized_latency = datawidth_df[metric] / reference_value * 100
        
        # Append values to the lists
        x_values.append(normalized_latency)
        y_values.append(accuracy)
        c_values.append(j)
        
        # Store the normalized value in the dictionary
        normalized_values_dict[design_setting][datawidth_setting] = normalized_latency.values[0]
    
    # Plot scatter subplot
    axes[i].scatter(x_values, y_values, s=70, alpha=0.75, c=c_values, cmap=cmap)
    axes[i].set_xlabel("Normalized " + metric + " [%]", fontsize=16, fontweight='bold')
    axes[i].set_title(design_setting, fontsize=16)    
    
    axes[i].tick_params(axis='x', which='major', labelsize=14)
    axes[i].tick_params(axis='y', which='major', labelsize=14)
    axes[i].grid(True, color='lightgray', linestyle='-', linewidth=0.5, alpha=0.25)
    axes[i].set_ylim([0, 100])  # Set the y-axis range
    axes[i].set_yticks(range(0, 101, 10))  # Set the y-axis ticks
    axes[i].set_xlim([0, 105])  # Set the y-axis range
    
# Print the dictionary with the normalized values
print(normalized_values_dict)

# Set y-axis label for the leftmost subplot
axes[0].set_ylabel("Accuracy [%]", fontsize=16, fontweight='bold')

# Create a colorbar within the same plot
cbar_ax = fig.add_axes([0.175, -0.05, 0.7, 0.03])  # Specify the position of the colorbar
cbar = fig.colorbar(axes[0].collections[0], cax=cbar_ax, cmap=cmap, orientation='horizontal', ticks=range(len(datawidth_settings)))
cbar.set_ticks(range(len(datawidth_settings)))
cbar.set_ticklabels(sorted(list(datawidth_settings)), fontsize=14)
cbar.set_label("Datawidth Setting", fontsize=16)


# Save the plot
metric_name = metric.replace(" ", "_")
plt.tight_layout()
fig_location = "plots/Experiment3"
if not os.path.exists(fig_location):
    os.makedirs(fig_location)
plt.savefig(os.path.join(fig_location, f'{arch_name}_relative_accuracy_{metric_name}_scatterplot_quantized_MobileNet.png'), bbox_inches='tight')
plt.savefig(os.path.join(fig_location, f'{arch_name}_relative_accuracy_{metric_name}_scatterplot_quantized_MobileNet.pdf'), bbox_inches='tight')


plt.show()
```
